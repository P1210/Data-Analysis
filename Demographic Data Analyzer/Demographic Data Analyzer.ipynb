### Q2) What is the average age of men?

```python
df=data.loc[data['sex'] != 'Female']
df['age'].mean()
```

```python
df1=data[['education','education-num']]
df1.value_counts()
```

### Q3) What is the percentage of people who have a Bachelor's degree?

```python
df2 = df1[df1['education-num'].isin(['13']) == True]
df2.shape
```

```python
percent = (df2.shape[0]/data.shape[0])*100
print(percent)
```

### Q4) What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?

```python
df3 = data[data['education-num'].isin(['13','14','16']) == True]
df3.shape
```

```python
df3.head(7)
```

```python
df3['salary'].value_counts()
```

```python
percent = (df3['salary'].value_counts()[1]/data.shape[0])*100
print(percent)
```

### Q5) What percentage of people without advanced education make more than 50K?
